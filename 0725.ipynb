반복문

```python
#while True인동안:
카운트 = 0
while 카운트<10:
  print(카운트)
  카운트 = 카운트 + 1
```

```python
#while True인동안:
카운트 = 0
while 카운트<10:
  카운트 = 카운트 + 1
  if (카운트==1):
    continue

  if (카운트%2==1):
    print(카운트)
```
